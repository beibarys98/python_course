```python
import random

x = 5
y = 5
score = 0
x2= 8
y2 =7
while True:
    for i in range(1,12):
        for j in range(1,12):  
            if i==y and j == x:
                print(' @ ',end='')
            elif i==y2 and j==x2:
                print(' $ ',end='')
            elif j == 1 or j == 11:
                print(' | ',end='')
            elif i == 1 or i == 11:
                print('___',end='')
            else:
                print(' . ', end='')
        print('')

        
    print('score: ',score)
    button= input()
    if button== 'w':
        y =y-1
        if y<2:
            y = 10
    if button =='s':
        y = y+1
        if y>10:
            y = 2
    if button == 'd':
        x =x+1
        if x > 10:
            x = 2
    if button == 'a':
        x =x-1
        if x < 2:
            x = 10
    if button == '!':
        break
    if y == y2 and x==x2:
        score+=1
        x2= random.randint(2,10)
        y2 =random.randint(2,10)
```
